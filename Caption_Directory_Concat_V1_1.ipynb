# SETUP PATHS HERE:

```python
import os

# Define the root directory to scan
root_dir = "/content/drive/MyDrive/PATH"  # @param {type:'string'}

# Define the output directory for the new text files
output_dir = "/content/drive/MyDrive/PATH" # @param {type:'string'}
```

# Run Caption Directory Concatenator

```python
# Iterate over each directory in the root directory
for dirpath, dirnames, filenames in os.walk(root_dir):
    print(f"Scanning directory: {dirpath}")
    
    # Create a list of all .txt files in the current directory
    txt_files = [f for f in filenames if f.endswith('.txt')]

    # If there are no .txt files in the directory, skip to the next one
    if len(txt_files) == 0:
        print("No .txt files found in this directory.")
        continue
    
    # Create a new text file with the same name as the current directory
    output_filename = os.path.join(output_dir, os.path.basename(dirpath) + '.txt')
    print(f"Creating output file: {output_filename}")
    with open(output_filename, 'a') as output_file:
        
        # Iterate over each .txt file in the current directory and add its first line to the output file
        for txt_file in txt_files:
            txt_file_path = os.path.join(dirpath, txt_file)
            print(f"Adding first line of {txt_file} to output file.")
            with open(txt_file_path, 'r') as txt_file_obj:
                first_line = txt_file_obj.readline()
                output_file.write("\n" + first_line.strip()) # Add the first line on a new line
                
print("Done scanning directories.")  
```
